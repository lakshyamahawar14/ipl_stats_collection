```python
import pandas as pd

df = pd.read_csv('./ipl_2024_players_all_stats.csv')

df.head()
```

```python
team_colors = {
    'PBKS' : 'D50F28',
    'DD' : '021346',
    'CSK' : 'EED201',
    'GL' : 'E25627',
    'MI' : '1230D6',
    'RCB' : 'C01F23',
    'KTK' : 'EB4917',
    'KKR' : '452858',
    'PWI' : '387A95',
    'DC' : '2937CE',
    'GT' : '3C4A63',
    'RPS' : 'DF4AAB',
    'RR' : 'DF3C65',
    'LSG' : '3C4874',
    'DEC' : '354D5E',
    'SRH' : 'FF5131',
}

import matplotlib.colors as mcolors


colors = {}
for team, hex_color in team_colors.items():
    rgba_color = mcolors.to_rgba('#' + hex_color)
    colors[team] = rgba_color

colors
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Drop rows where 'Runs' is 'No stats' and convert 'Runs' to numeric type
df = df[df['Runs'] != 'No stats']
df['Runs'] = pd.to_numeric(df['Runs'], errors='coerce')

# Filter only the rows where 'Year' is 'Career'
career_runs_df = df[df['Year'] == 'Career']

# Sort the DataFrame by 'Runs' column in descending order and select the top 10 players
top_10_players = career_runs_df.sort_values(by='Runs', ascending=False).head(10)

# Plot the top 10 players with their career runs
plt.figure(figsize=(12, 8))
plt.bar(top_10_players['Player Name'], top_10_players['Runs'], color=[colors[team] for team in top_10_players['Team Name']])
plt.xlabel('Player Name')
plt.ylabel('Total Runs')
plt.title('Top 10 Players with Most Career Runs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Filter only the rows where 'Year' is 'Career'
career_wickets_df = df[df['Year'] == 'Career'].copy()

# Convert 'Wickets' to numeric type
career_wickets_df.loc[:, 'Wickets'] = pd.to_numeric(career_wickets_df['Wickets'], errors='coerce')

# Sort the DataFrame by 'Wickets' column in descending order and select the top 10 players
top_10_bowlers = career_wickets_df.sort_values(by='Wickets', ascending=False).head(10)

# Plot the top 10 bowlers with their career wickets
plt.figure(figsize=(12, 8))
plt.bar(top_10_bowlers['Player Name'], top_10_bowlers['Wickets'], color=[colors[team] for team in top_10_bowlers['Team Name']])
plt.xlabel('Player Name')
plt.ylabel('Total Wickets')
plt.title('Top 10 Bowlers with Most Career Wickets')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Filter only the rows where 'Year' is 'Career'
career_not_outs_df = df[df['Year'] == 'Career'].copy()

# Convert 'Not Outs' to numeric type
career_not_outs_df['Not Outs'] = pd.to_numeric(career_not_outs_df['Not Outs'], errors='coerce')

# Sort the DataFrame by 'Not Outs' column in descending order and select the top 10 players
top_10_not_outs = career_not_outs_df.sort_values(by='Not Outs', ascending=False).head(10)

# Plot the top 10 players with their total not outs
plt.figure(figsize=(12, 8))
plt.bar(top_10_not_outs['Player Name'], top_10_not_outs['Not Outs'], color=[colors[team] for team in top_10_not_outs['Team Name']])
plt.xlabel('Player Name')
plt.ylabel('Total Not Outs')
plt.title('Top 10 Players with Most Career Not Outs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
